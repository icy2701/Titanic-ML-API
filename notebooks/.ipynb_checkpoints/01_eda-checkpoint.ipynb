```python
#Load the data

import pandas as pd
df=pd.read_csv('../data/train.csv') #reads the CSV file and converts it into a table
df.head()  # shows first 5 rows
```

```python
#Understand the data

df.info()   # column names, types, non-null counts
df.describe()  # stats like mean, min, max
```

```python
#Check missing values

df.isnull().sum()

#df.isnull() → checks every cell, returns True if missing, False if not
#.sum() → counts all the True values per column
```

```python
#Survival rate by Sex, Pclass, Embarked

import matplotlib.pyplot as plt

# Survival by gender
print(df.groupby('Sex')['Survived'].mean())

#Survival by Pclass
print(df.groupby('Pclass')['Survived'].mean())

#Survival by where they boarded
print(df.groupby('Embarked')['Survived'].mean())

#df.groupby('Sex') → groups all rows by gender (male/female)
#['Survived'] → looks at only the Survived column
#.mean() → calculates average survival (since it's 0 or 1, mean = survival rate)
```

```python
#Age Distrubution

df['Age'].hist(bins=20)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution')
plt.show()


#df['Age'] → selects only the Age column
#.hist(bins=30) → draws a histogram with 30 bars
#bins=30 → divides the age range into 30 groups
#plt.xlabel/ylabel/title → adds labels to the chart
#plt.show() → displays the chart
```
