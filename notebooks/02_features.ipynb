```python
import pandas as pd

df = pd.read_csv('../data/train.csv')
```

```python
#Fill Missing Age Values

df['Age'].fillna(df['Age'].median(),inplace=True)

#PartMeaningdf['Age'] -> Select the Age column
#.fillna() -> Fill all NaN (missing) values with something
#df['Age'].median() -> Calculate the middle value of Age (28.0)
#inplace=True -> Change the original df directly, don't create a copy

#Why Median and not Mean??
#Mean(AVG) gets affected by extreme values (very old passengers) - sensitive to outliers
#Median is more reliable — it's just the middle number - robust to outliers
#Example: Ages [10, 20, 30, 80] → mean=35, median=25 (20+30/2).

print(df['Age'].isnull().sum())  #verify should be 0
```

```python
#Encode Sex Column

df['Sex'] = df['Sex'].map({'male': 0 , 'female' : 1})

#.map({}) -> Replace values using a dictionary
#ML models only understand numbers, not text

print(df['Sex'].unique())  # verify should show [0, 1]
```

```python
#Fill Missing Embarked Values

df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

#df['Embarked'].mode()  ->Find the most frequent value
#[0] -> Take the first result (mode returns a list)
#Mode = most frequently occurring value
#Embarked has S, C, Q → S appears most often
#So we fill the 2 missing values with 'S'

print(df['Embarked'].isnull().sum()) # verify should be 0
```

```python
#One-Hot Encode Embarked

df=pd.get_dummies(df, columns=['Embarked'])

#pd.get_dummies() -> Convert text categories into multiple number columns
#Sex only has 2 values → simple 0/1 works
#Embarked has 3 values (S, C, Q) → if we do S=0, C=1, Q=2 the model thinks Q > C > S which is wrong!
#One-hot encoding creates separate columns for each value -> Embarked_C  Embarked_Q  Embarked_S

print(df.columns.tolist())
```

```python
#Drop Useless Columns

df.drop(['PassengerId','Name','Ticket','Cabin'], axis=1 , inplace=True)

#axis=1  -> 1 means columns, 0 means rows

#Name -> Every name is unique, no pattern to learn
#Ticket -> Random alphanumeric, no useful pattern
#Cabin -> 687 out of 891 missing — too unreliable
#PassengerId -> Just a serial number, meaningless
```

```python
print(df.shape)
print(df.columns.tolist())
```

```python
#Create FamilySize Feature

df['FamilySize']= df['SibSp']+df['Parch']+1

#+ 1 -> Add the passenger themselves
#df['FamilySize'] -> Store result as a new column
#SibSp=1, Parch=2 → FamilySize = 1+2+1 = 4

print(df[['SibSp','Parch','FamilySize']].head(10))
```

```python
#Save Processed Data

df.to_csv('../data/processed.csv',index=False)

#df.to_csv() -> Save DataFrame as a CSV file
#'../data/processed.csv' -> Save it in the data/ folder
#index=False -> Don't save the row numbers (0,1,2...) as a column
```
